```python
import pandas as pd
restaurant_df = pd.read_csv('eat_testdata.csv')
display(restaurant_df)
```

```python
m = folium.Map(location=[25.017311960327163, 121.53973532014066], tiles="OpenStreetMap", zoom_start=16)
# 加入餐廳點位
for i in range(0,len(restaurant_df)):
    html=f"""
        <h1> {restaurant_df.iloc[i]['name']}</h1>
        <p>{restaurant_df.iloc[i]['type']}</p>
        <ul>
            <li>地址：{restaurant_df.iloc[i]['address']}</li>
            <li>電話：{restaurant_df.iloc[i]['phone']}</li>
        </ul>
        </p>
        <p>點擊 <a href={restaurant_df.iloc[i]['map']} target="_blank">進入地圖</a></p>
        """
    iframe = folium.IFrame(html=html, width=500, height=250)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        location=[restaurant_df.iloc[i]['lat'], restaurant_df.iloc[i]['lon']],
        popup=popup
    ).add_to(m)
m
```

使用説明：點擊藍色marker即可顯示餐廳資訊，點擊底下文字跳轉至googlemap頁面。

*p/s. 互動式地圖在html檔或nbviewer中查看*

```python
m.save('eat_testmap.html') # 儲存為html檔
```
