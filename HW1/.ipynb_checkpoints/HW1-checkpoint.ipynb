```python
import pandas as pd

# 讀取 CSV 檔案並轉換為 DataFrame
df = pd.read_csv('qs20rank2324.csv')
print("23/24年世界排名前20的大學:")
display(df)
print()

# 將 DataFrame 轉換為 Set
top20_2023 = set(df['2023'])
top20_2024 = set(df['2024'])

# 聯集
union_set = top20_2023.union(top20_2024)
print("聯集:", union_set) # 所有學校

print()

# 交集
intersection_set = top20_2023.intersection(top20_2024)
print("交集:", intersection_set) # 穩定上榜

print()
# 差集
difference23_set = top20_2023.difference(top20_2024)
print("差集（23年榜上）:", difference23_set) # 23上榜24跌落神壇
difference24_set = top20_2024.difference(top20_2023)
print("差集（24年榜上）:", difference24_set) # 24年追上
```
